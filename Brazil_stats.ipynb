# Mato Grosso

```python
def extract_ICUs(date, links, important_links):
    """Extracts data from the secretaria de saude of Mato Grosso
    Input:
        links: (array) array with the links
        important_links: (int) how many days to retrieve, counting from the present to the past. 
        date: starting date
    Output:
        all_data: (df) with all the ICUs of MT. 
    """
    day = 0
    df_list = []
    for link in trange(0, important_links):
        pdf_path = links[link]

        # get state data for the day
        MT_data = tabula.read_pdf(pdf_path, pages=3, columns = [48, 114, 276, 348, 432], guess=True)
        state_data = MT_data[0]
        state_data.columns = ['Municipio', 'Unidade', 'Leitos_UTI', 'Total_Casos', 'Ocupao']
        state_data = state_data[2:]
        state_data['date'] = date - pd.DateOffset(day)
        df_list.append(state_data)

        # get municipal data for the day
        municipal_data = MT_data[1]
        municipal_data.columns = ['Municipio', 'Unidade', 'Leitos_UTI', 'Total_Casos', 'Ocupao']
        municipal_data = municipal_data[2:]
        municipal_data['date'] = date - pd.DateOffset(day)
        df_list.append(municipal_data)

        day += 1

        
    all_data = pd.concat(df_list)
    
    # some processing 
    all_data['Leitos_UTI'] = all_data['Leitos_UTI'].str.replace('*', '').astype(int)
    all_data['Total_Casos'] = all_data['Total_Casos'].str.replace('*', '').astype(int)
    
    return all_data
```

```python
# Get all links to corona reports
url = "http://www.saude.mt.gov.br/informe/584"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
req = requests.get(url=url, headers=headers)
soup = BeautifulSoup(req.content, "html.parser")

links = []
for link in soup.findAll('a', attrs={'href': re.compile("^http://www.saude.mt.gov.br/arquivo/")}):
    links.append(link.get('href'))
```

```python
important_links = 8 # the last eight days
date = pd.to_datetime('2020-06-27')
all_data = extract_ICUs(date,links, important_links)
```

```python
# Filter by Grande Cuiab
grande_cuiaba_filt = (all_data['Municipio'] == 'Cuiab') | (all_data['Municipio'] == 'Vrzea Grande')
grande_Cuiaba = all_data.loc[grande_cuiaba_filt]
# Calculate available ICUs
summary = grande_Cuiaba.groupby(['date']).agg({'Leitos_UTI': sum, 'Total_Casos':sum}).reset_index()
summary['Ocupao'] = summary['Total_Casos'] / summary['Leitos_UTI'] * 100
```

```python
# Plot the status of the ICUs
summary.iplot(
            kind = 'bar',
#                     barmode = 'stack',
            opacity=0.8,
            size=1,
            symbol=8,
            xTitle='Data',
            yTitle='Ocupao',
            title='Ocupao UTIs na Grande Cuiab',
#                             fill = True
            x = 'date',
            y = 'Ocupao'
            )
```

```python
lista_muni.drop(columns={'Unnamed: 0'}, inplace=True)
lista_muni.head()
```

```python
lista_muni['# Casos'] = lista_muni['# Casos'].astype('float32')
lista_muni['bitos'] = lista_muni['bitos'].astype('float32')
lista_muni['# Casos'] = lista_muni['# Casos'].astype('str')
lista_muni['bitos'] = lista_muni['bitos'].astype('str')
```

```python
def delete_point(number):
    index = number.find('.')
    if len(number[(index+1):])>=3:
        number = number.replace('.','')
    return number
```

```python
lista_muni['Casos_2'] = lista_muni.apply(lambda x: delete_point(x['# Casos']), axis=1).astype('float32')
lista_muni['Obitos_2'] = lista_muni.apply(lambda x: delete_point(x['bitos']), axis=1).astype('float32')
```
